```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a Horizontal Bar Chart

# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
# To invert the y-axis to have the months in ascending order, use the gca() method. 
# The gca() method means "get current axes." We can chain the gca() method to the invert_yaxis() method by using gca().invert_yaxis()
plt.barh(x_axis, y_axis)
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label='New York')
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color='cyan', label='Chicago')
ax.invert_yaxis()
ax.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")

```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis,color='r', label='Chicago')
plt.legend()
plt.invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,color='red', label='Chicago')
ax.invert_yaxis()
ax.legend()
plt.xlim(0, 50)
plt.ylabel("Date")
plt.xlabel("Fare($)")
```

```python
#  Bubble Chart
plt.scatter(x_axis, y_axis)
```

```python
#  Bubble Chart with size parameter
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# list comprehension 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis,color= s = [i * 5 for i in y_axis])
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis)


# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
